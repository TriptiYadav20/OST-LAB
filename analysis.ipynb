```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Load sample dataset
# If data is small, you can put it in repo/data/ folder
df = pd.DataFrame({
    "Name": ["A", "B", "C", "D"],
    "Marks": [85, 67, 90, 76]
})

# Step 3: Display data
print(df)

# Step 4: Plot simple graph
plt.plot(df["Name"], df["Marks"], marker='o')
plt.title("Student Marks")
plt.xlabel("Name")
plt.ylabel("Marks")
plt.show()
```
